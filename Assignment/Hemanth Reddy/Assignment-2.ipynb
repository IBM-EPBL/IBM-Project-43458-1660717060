## Finding and Replacing Outliers

```python
sns.boxplot(x=data.CreditScore)
```

```python
q1 = data.CreditScore.quantile(0.25)
q3 = data.CreditScore.quantile(0.75)
IQR = q3 - q1
lower_limit = q1 - 1.5 * IQR
```

```python
data.median(numeric_only=True)
```

```python
data['CreditScore'] = np.where(data['CreditScore'] < lower_limit, 7, data['CreditScore'])
```

```python
sns.boxplot(x=data.CreditScore)
```

## Encoding Categorical columns


```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
data.Gender = le.fit_transform(data.Gender)
data.Geography = le.fit_transform(data.Geography)

data.head()
```

## Splitting into Independent(X) and Dependent(Y) Variable


```python
Y = data['Surname']
Y
```

```python
X = data.drop(columns=['Surname'], axis=1)
X.head()
```

```python
from sklearn.preprocessing import scale

X_Scaled = pd.DataFrame(scale(X), columns=X.columns)
X_Scaled.head()
```

```python
from sklearn.model_selection import train_test_split
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X_Scaled, Y, test_size=0.3, random_state=0)
```
